## **This script is used to create precision-recall curves**

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# models used
models = ['VGG16', 'KNMI']
```

```python
# assign night data to variables
p1_pm = np.array([0.048525        , 0.328502, 1.        ])
r1_pm = np.array([1.        , 0.888889, 0.        ])
auc1_pm = 0.61139
p2_pm = np.array([0.04988042        , 0.50638298, 1.        ])
r2_pm = np.array([1.        , 0.81506849, 0.        ])
auc2_pm = 0.66533
```

```python
# plot roc curves for each model
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7))
plt.title('Precision-Recall Curve: Night Images', fontsize = 16)
ax.plot(r1_pm, p1_pm, 'tab:blue', label = 'VGG16  = %0.2f' % auc1_pm)  
ax.plot(r2_pm, p2_pm, 'tab:orange', label = 'KNMI = %0.2f' % auc2_pm)

plt.legend(loc = 'upper right', fontsize = 14)
#plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1])
plt.ylim([0.0, 1])
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylabel('Precision', fontsize = 16)
plt.xlabel('Recall', fontsize = 16)
#plt.show()
plt.savefig('/home/ubuntu/michael/model_output/prec_rec_pm2.png')
```
