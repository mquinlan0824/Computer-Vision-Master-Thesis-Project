## **Store day/fog test images in unbalanced_day_test_fog directory**

```python
import os
import shutil
```

```python
# retrieve images from source directory
src = '/home/ubuntu/backup_images_2/images01062021' 
```

```python
# create a funtion to iterate through list of files and copy to appropriate dst_dir

def copyFiles(fn_list, src_dir, dst_dir):
    for filename in fn_list:
        #print(filename)
        path = os.path.join(src_dir, filename)
        #print(path)
        if os.path.exists(path):
            shutil.copyfile(path, dst_dir + '/' + filename)
        else:
            continue
    return
```
