# **ROC Curves for Model Performance comparison**

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# models used
models = ['VGG16_3Block', 'Dropout', 'Augmentation', 'VGG16_3Block_DropAug', 'Inception', 'Resnet50', 'EfficientNetB0']
```

```python
models = ['VGG16', 'KNMI']
```

## **Plot ROC curves for Night images only**

```python
# data for night images
fp_pm = [9, 10, 11, 9, 7, 8, 21]
tp_pm = [132, 132, 131, 136, 129, 126, 125]
fn_pm = [21, 21, 22, 17, 24, 27, 28]
tn_pm = [129, 128, 131, 129, 131, 130, 117]
auc_pm = [0.8988, 0.8951, 0.8883, 0.9118, 0.8962, 0.8828, 0.8324]
```

```python
test_fp = [278, 89]
test_tp = [136, 114]
test_fn = [17, 32]
test_tn = [2722, 2692]
auc = [0.8981, 0.8744]
```

```python
# night data store values in a dataframe
test = pd.DataFrame()
test['Classifier'] = models
test['fp'] = test_fp
test['tp'] = test_tp
test['fn'] = test_fn
test['tn'] = test_tn
test['auc'] = auc
test['tpr'] = test.tp / (test.tp + test.fn)
test['fpr'] = test.fp / (test.fp + test.tn)
test
```

```python
fpr_vgg = np.array([0.        , 0.092667, 1.        ])
tpr_vgg = np.array([0.        , 0.888889, 1.        ])
auc_vgg = 0.8981

fpr_knmi = np.array([0.        , 0.032003, 1.        ])
tpr_knmi = np.array([0.        , 0.780822, 1.        ])
auc_knmi = 0.8744
```

```python
# plot roc curves for each model
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7))
plt.title('Receiver Operating Curve: Test Set', fontsize = 16)
ax.plot(fpr_vgg, tpr_vgg, 'tab:blue', label = 'VGG16 = %0.2f' % auc_vgg)
ax.plot(fpr_knmi, tpr_knmi, 'tab:orange', label = 'KNMI = %0.2f' % auc_knmi)

plt.legend(loc = 'lower right', fontsize = 14)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylabel('True Positive Rate', fontsize = 16)
plt.xlabel('False Positive Rate', fontsize = 16)
#plt.show()
plt.savefig('/home/ubuntu/michael/model_output/roc_curve_test.png')
```
